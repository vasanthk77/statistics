```python
sample_mean={}
mean=[]
for i in range(0,100):
  samples=random.sample(range(0,df.shape[0]),50)
  mean.append(df.iloc[samples,0].mean())
sample_mean['100_samples_with_50_values']=mean
```

```python
np.mean(sample_mean['100_samples_with_50_values'])
```

```python
sns.displot(sample_mean['100_samples_with_50_values'],kde=True)
```
